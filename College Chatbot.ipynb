```python
# JSON file se data load karna
def load_data():
    with open("rits_chatbot_data.json", "r") as json_file:
        return json.load(json_file)

data = load_data()
```

```python
# Chatbot function
def chatbot(user_input):
    for category, details in data.items():
        for pattern in details["patterns"]:
            if pattern.lower() in user_input.lower():
                return details["responses"][0]
    return "Sorry, mujhe samajh nahi aaya. Please RITS ke kisi specific topic ke baare mein puchhein."
```

```python
# User interaction
while True:
    user_input = input("Aap kya puchhna chahte hain? (exit likhne par band hoga): ")
    if user_input.lower() == "exit":
        break
    response = chatbot(user_input)
    print("Chatbot:", response)
```
